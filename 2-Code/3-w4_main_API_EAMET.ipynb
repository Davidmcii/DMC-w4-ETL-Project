```python
ini_time='2020-01-02'
end_time='2022-08-01'
reg_22_20=[]

for indicativo in estaciones.indicativo:
    station=indicativo
    querry='api/valores/climatologicos/diarios/datos/fechaini/{}T00:00:00UTC/fechafin/{}T23:59:59UTC/estacion/{}'.format(ini_time,end_time,station)

    res_aemet = req.get(url+querry+TOKEN, headers=headers)
    
    result= res_aemet.json()
    if 'exito' in str(result):
        url_datos=result['datos']
        conn=req.get(url_datos)
        try:
            reg_22_20=reg_22_20+conn.json()
            print(indicativo, 'added')
        except:
            print(indicativo, 'ERROR JSON')
    else:
        print('ERROR en ',indicativo, 'skiped')


reg_22_20
    
```
