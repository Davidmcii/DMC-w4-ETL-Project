```python
villages=pd.read_csv('./111(3).csv', sep=';')



villages
```

```python
villages.rename(columns={'Población':'MUNICIPIO'}, inplace=True)

villages['complete_name']=villages.Provincia + ' ' + villages.MUNICIPIO

villages.head()
```

```python
for ind, e in enumerate(villages.Población):
    villages.Provincia[ind]= villages.Provincia[ind].upper()
    if '('in e:
        villages.Población[ind]= villages.Población[ind].replace('(','')
        villages.Población[ind]= villages.Población[ind].replace(')','')



villages
```

```python
pueblos=[]
no_encon=[]
latitude=[]
longitude=[]

for i in range(len(FV.MUNICIPIO)):              
    busqueda=FV.MUNICIPIO[i]
    flag=False
    
    for j, el in enumerate(villages.MUNICIPIO):
        if busqueda==el and FV.PROVINCIA[i]==villages.Provincia[j]:
            pueblos.append(el)
            latitude.append(villages.Latitud[j])
            longitude.append(villages.Longitud[j])
            flag=True

    if flag==False:
        no_encon.append(FV.MUNICIPIO[i])
        pueblos.append(None)
        latitude.append(None)
        longitude.append(None)
        
        
pueblos
```

```python
location=[]
latitude=[]
longitude=[]
no_encontrados=[]
encontrados=[]

for el in list_no:
    geolocator = Nominatim(user_agent="Your_Name")
    city =el
    country ="Spain"
    loc = geolocator.geocode(city+','+ country)
    try:
        print(el, " + " ,loc.address)
        location.append(loc.address)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
        encontrados.append(el)
    except:
        print('no encontrado')
        no_encontrados.append(el)
    time.sleep(3)
```
