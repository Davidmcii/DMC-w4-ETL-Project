```python
url= 'https://aeeolica.org/sobre-la-eolica/mapa-de-parques-eolicos/'

urls_ccaa=[]
for comuni in web2:

    wind=url+str(comuni)
    urls_ccaa.append(wind)
    
urls_ccaa
```

```python
%%time

info_parques=[]

for ccaa in urls_ccaa:
    driver.get(ccaa)
    caja=driver.find_element(By.XPATH,'//*[@id="post-4687"]/div/div/div/div[3]/div')
    parque_ccaa= [e.text for e in caja.find_elements(By.TAG_NAME, 'li')]
    #parque_ccaa.pop(1)
    #parque_ccaa.pop(0)
    for name in parque_ccaa:
        name2=name.removesuffix('\n+ Info').replace('Provincia: ','').replace('Término Municipal: ','').replace('Sociedad promotora: ','').replace('Potencia MW: ','').replace(',','.')
        name3=name2.split('\n',4)
        info_parques.append(name3)

info_parques
```

```python
PPEE_esp=pd.DataFrame(info_parques)

```

```python
PPEE_esp.to_csv('./ppee_esp.csv', index=False, sep =',')
```
