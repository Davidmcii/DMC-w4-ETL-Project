```python
PPEE=pd.read_csv('./ppee_esp.csv')

PPEE
```

```python
PPEE.rename(columns = {'0': 'ZERO'}, inplace = True)
PPEE.rename(columns = {'1': 'PROVINCIA'}, inplace = True)
PPEE.rename(columns = {'2': 'MUNICIPIO'}, inplace = True)
PPEE.rename(columns = {'3': 'SOCIEDAD'}, inplace = True)
PPEE.rename(columns = {'4': 'POTENCIA'}, inplace = True)

PPEE
```

```python
municipios=[]
for i, el in enumerate(PPEE.MUNICIPIO):
    
    if ' y ' in PPEE.MUNICIPIO[i]:

        muni=PPEE.MUNICIPIO[i].split(' y ')
        muni=muni[1]
        municipios.append(muni)
    
    else:
        municipios.append(el)
        
    
municipios

PPEE['municipios2']=municipios

PPEE['nombre_completo']=PPEE.municipios2+' '+ PPEE.PROVINCIA

PPEE
```

```python
set_no=set(PPEE.nombre_completo)

len(set_no)
```

```python
location=[]
latitude=[]
longitude=[]
no_encontrados=[]
encontrados=[]

for el in set_no:
    geolocator = Nominatim(user_agent="Your_Name")
    city =el
    country ="Spain"
    loc = geolocator.geocode(city+','+ country)
    try:
        print(el, " + " ,loc.address)
        location.append(loc.address)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
        encontrados.append(el)
        
    except:
        print('no encontrado')
        no_encontrados.append(el)
        
    time.sleep(3)
```
