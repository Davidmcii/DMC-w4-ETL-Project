```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By

from IPython.display import Image

#driver configuration
opciones=Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
opciones.add_argument('--incognito')

driver = ChromeDriverManager().install() 
driver = webdriver.Chrome(driver,options = opciones)

```

```python
csp1= 'https://solarpaces.nrel.gov/by-country/ES'


html=req.get(csp1).content

soup=bs(html, 'html.parser')

soup
```

```python
tarjetas=soup.find_all('div', class_='views-field views-field-title')

webs=[]
for i in range(len(tarjetas)):

    ref=tarjetas[i].find('a')

    result=ref.attrs['href']
    urls='https://solarpaces.nrel.gov'+result
    webs.append(urls)


webs

```

```python
CSP_esp=[]

for el in webs:

    csp_pro= el

    html=req.get(csp_pro).content

    soup=bs(html, 'html.parser')

    tarjetas=soup.find_all('div', class_='row')

    llave=[]
    contenido=[]

    i=4
    
    tarjetas2=tarjetas[i].find_all('td')

    for j in range(len(tarjetas2)):
        
        if 'Remuneration EUR' not in tarjetas2[j].text.strip():
            contenido.append(tarjetas2[j].text.strip())
        
    tarjetas3=tarjetas[i].find_all('th')

    for k in range(len(tarjetas3)):
        llave.append(tarjetas3[k].text.strip())
    
    dictio=dict(zip(llave,contenido))

    CSP_esp.append(dictio)

CSP_esp
```

```python
import json

with open('CSP_esp.json', 'w') as json_file:
    json.dump(CSP_esp, json_file)
```
